```python
from sklearn.linear_model import Lasso
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
def find_best_model(X,y):
    algos = {
        'linear_regression' : {
            'model' : LinearRegression(),
            'params' : {}
        },
        'lasso' : {
            'model' : Lasso(),
            'params' : {
                'alpha' : [1,2],
                'selection' : ['random', 'cyclic']
            }
        },
        'decision_tree' : {
            'model' : DecisionTreeRegressor(),
            'params' : {
                'criterion' : ['mse', 'friedman_mse'],
                'splitter' : ['best', 'random']
            }
        },
        'random_forest' : {
            'model' : RandomForestRegressor(),
            'params' : {
                'n_estimators' : [1,5,10]
            }
        }
        
    }

    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gcv = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gcv.fit(X, y)
        scores.append({
            'model' : algo_name,
            'best_score' : gcv.best_score_,
            'best_param' : gcv.best_params_
        })
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_param'])
```

```python
find_best_model(X, y)
```

```python
def predict_price(vehicle_age, km_driven, mileage, engine, max_power, seats, brand, model, seller_type, fuel_type, transmission_type):
    brand_loc = np.where(X.columns == brand)[0][0]
    model_loc = np.where(X.columns == model)[0][0]
    seller_type_loc = np.where(X.columns == seller_type)[0][0]
    fuel_type_loc = np.where(X.columns == fuel_type)[0][0]
    transmission_type_loc = np.where(X.columns == transmission_type)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = vehicle_age
    x[1] = km_driven
    x[2] = mileage
    x[3] = engine
    x[4] = max_power
    x[5] = seats
    
    if brand_loc>=0:
        x[brand_loc] = 1
    if model_loc>=0:
        x[model_loc] = 1
    if seller_type_loc>=0:
        x[seller_type_loc] = 1
    if fuel_type_loc>=0:
        x[fuel_type_loc] = 1
    if transmission_type_loc>=0:
        x[transmission_type_loc] = 1

    return rfc_model.predict([x])[0]
```

```python
predict_price(1, 2000, 19.9, 1197, 200, 2, 'Hyundai', 'Grand', 'Individual', 'Petrol', 'Manual')
```
